# Haygood Model

Regarding trangene containment strategies, when it comes to predicting mean escape rates, the Haygood model is a resonable one, such that the following general form for mean escape rate, $\mu$, is a valuable one:
$$
\mu = \frac{s'}{m(1-s')rN(1-e^{-2s})}
$$
where\
$\mu$ is the mean escape rate for the respective transgene\
$s'$ is a selection coefficient (almost like a measure of relative fitness, in this case the opposite) against first generation hybrids, such that $0 << s' < 1\
$s$ is a selection coefficient for transgenic recombinants, such that $0 < s$\
$m$ is the fraction of pollen in the population that is indeed effective, such that $0 < m << 1\
$r$ is the frequency of recombination between the transgene and the mitigation gene (if there is one), such that 0 < r << 0.5\
and $N$ is the number of effective plants in the wild-type population

```python
import numpy as np, matplotlib.pyplot as plt


# Arbitrary choice of parameter's values
num_gen = np.arange(0,100,1) # Number of generations to run the simulation over
s_prime = 0.5 * (1 - (0.4 / num_gen)) # Perhaps other parameters would change with generation's increase, however these ones seem the most obvious
initial_s_prime = 0.5
s = 0.6 * (1 + (5 / num_gen))# Perhaps other parameters would change with generation's increase, however these one seem the most obvious
initial_s = 0.6
m = 0.2
r = 0.1
N = 30
initial_m_e_r = (initial_s_prime)/(m*(1-initial_s_prime)*r*N*(1-np.e**(-2*initial_s)))
m_e_r = (s_prime)/(m*(1-s_prime)*r*N*(1-np.e**(-2*s)))
p_e = m_e_r / initial_m_e_r # Probability of escape


plt.figure()
plt.plot(num_gen, p_e, "k--")
plt.xlabel("NÂº Generations", size= 12,); plt.ylabel("$P_{\mu}$", size= 12)
plt.show()

```
